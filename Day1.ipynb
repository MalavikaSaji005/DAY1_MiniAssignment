```python
import os
import gradio as gr
from groq import Groqā

client = Groq(api_key=os.environ["GROQ_API_KEY"])

def chatbot(message, history, system_prompt, temperature):

    messages = [
        {
            "role": "system",
            "content": system_prompt
        }
    ]

    for msg in history:
        if msg["role"] != "system":
            messages.append(msg)

    messages.append(
        {
            "role": "user",
            "content": message
        }
    )

    response = client.chat.completions.create(
        model="llama-3.1-8b-instant",
        messages=messages,
        temperature=temperature,
        stream=True
    )

    partial_response = ""

    for chunk in response:
        content = chunk.choices[0].delta.content

        if content:
            partial_response += content
            yield partial_response


system_prompt_box = gr.Textbox(
    label="System Prompt",
    value="You are a football expert.",
    lines=4
)

temperature_slider = gr.Slider(
    minimum=0,
    maximum=2,
    value=0.7,
    step=0.1,
    label="Temperature"
)

demo = gr.ChatInterface(
    fn=chatbot,
    type="messages",
    additional_inputs=[
        system_prompt_box,
        temperature_slider
    ],
    title="Customizable AI Chatbot",
    description="Change the system prompt and temperature to control the chatbot's personality and creativity."
)

demo.launch(debug=True)
```
